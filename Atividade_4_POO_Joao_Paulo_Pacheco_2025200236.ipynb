```python
class Vector:
  def __init__(self, lenght, nums=None):
    if not isinstance(lenght, int) or lenght <= 0:
      raise TypeError("A dimensão deve ser um inteiro maior que zero.")
    isNumsNone = nums is None

    if not isNumsNone:
      if not isinstance(nums, (list, tuple)):
        raise TypeError("Os valores devem ser fornecidos em uma lista ou tupla.")
      if len(nums) != lenght:
        raise ValueError("A dimensão e o número de entradas fornecidas devem ser iguais.")
      self.elements = list(nums)
    else:
      self.elements = [0] * lenght
    self.length = lenght

  def get_dim(self):
    return self.length

  def get_entries(self):
    return tuple(self.elements)

  def __eq__(self, vec):
    if not isinstance(vec, Vector):
      raise TypeError("A comparação só pode ser feita com outro Vector.")
    if self.length != vec.get_dim():
      raise ValueError("Os vetores devem possuir o mesmo tamanho.")
    return self.length == vec.length and self.get_entries() == vec.get_entries()

  def __ne__(self, vec):
    if not isinstance(vec, Vector):
      raise TypeError("A comparação só pode ser feita com outro Vector.")
    if self.length != vec.get_dim():
      raise ValueError("Os vetores devem possuir o mesmo tamanho.")
    return not self == vec

  def __lt__(self, vec):
    if not isinstance(vec, Vector):
      raise TypeError("A comparação só pode ser feita com outro Vector.")
    if self.length != vec.get_dim():
      raise ValueError("Os vetores devem possuir o mesmo tamanho.")
    vec_elements = vec.elements
    for index,entry in enumerate(self.get_entries()):
      if entry >= vec_elements[index]:
        return False
    return True


  def __le__(self, vec):
    if not isinstance(vec, Vector):
      raise TypeError("A comparação só pode ser feita com outro Vector.")
    if self.length != vec.get_dim():
      raise ValueError("Os vetores devem possuir o mesmo tamanho.")
    vec_elements = vec.elements
    for index,entry in enumerate(self.get_entries()):
      if entry > vec_elements[index]:
        return False
    return True

  def __gt__(self, vec):
    if not isinstance(vec, Vector):
      raise TypeError("A comparação só pode ser feita com outro Vector.")
    if self.length != vec.get_dim():
      raise ValueError("Os vetores devem possuir o mesmo tamanho.")
    vec_elements = vec.elements
    for index,entry in enumerate(self.get_entries()):
      if entry <= vec_elements[index]:
        return False
    return True

  def __ge__(self, vec):
    if not isinstance(vec, Vector):
      raise TypeError("A comparação só pode ser feita com outro Vector.")
    if self.length != vec.get_dim():
      raise ValueError("Os vetores devem possuir o mesmo tamanho.")
    vec_elements = vec.elements
    for index,entry in enumerate(self.get_entries()):
      if entry < vec_elements[index]:
        return False
    return True

  def __add__(self, vec):
    if not isinstance(vec, Vector):
      raise TypeError("A adição só pode ser feita com outro Vector.")
    if self.length != vec.get_dim():
      raise ValueError("Os vetores devem possuir o mesmo tamanho.")
    vec_elements = vec.elements
    sum_list = list()
    for index,entry in enumerate(self.get_entries()):
      sum_list.append(entry + vec_elements[index])
    return Vector(self.length, sum_list).get_entries()

  def __sub__(self, vec):
    if not isinstance(vec, Vector):
      raise TypeError("A subtração só pode ser feita com outro Vector.")
    if self.length != vec.get_dim():
      raise ValueError("Os vetores devem possuir o mesmo tamanho.")
    vec_elements = vec.elements
    sub_list = list()
    for index,entry in enumerate(self.get_entries()):
      sub_list.append(entry - vec_elements[index])
    return Vector(self.length, sub_list).get_entries()

  def __mul__(self, scalar):
    if not isinstance(scalar, (int, float)):
      if isinstance(scalar, Vector):
        return self.dot(scalar)
      raise TypeError("O escalar deve ser um número (int ou float).")
    new_elements = [scalar * x for x in self.elements]
    return Vector(self.length, new_elements).get_entries()

  def dot(self, vec):
    if not isinstance(vec, Vector):
      raise TypeError("A multiplicação só pode ser feita com outro Vector.")
    if self.length != vec.get_dim():
      raise ValueError("Os vetores devem possuir o mesmo tamanho.")
    vec_elements = vec.elements
    inner_product = 0
    for index,entry in enumerate(self.get_entries()):
      inner_product += entry * vec_elements[index]
    return inner_product


v1 = Vector(3, [1, 2, 3])
v2 = Vector(3, [2, 2, 2])
v3 = Vector(3, [0, 0, 0])

print(v1 < v2)
print(v1 == Vector(3, [1,2,3]))
print(v2 > v3)
print(v3 - v2)
print(v1.dot(v2))
print(v1*4.5)
print(v1*v2)
```
